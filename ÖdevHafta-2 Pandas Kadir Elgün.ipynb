7-Her kıta için ve her alkol cinsi için ortalama alkol tüketimini gösteriniz.

8-Her kıta için ve aşağıdaki alkol cinsleri için belirtilen hesaplamaları gerçekleştiriniz.

beer_servings - std
spirit_servings - median 
wine_servings - min, mean, max
total_litres_of_pure_alcohol - mean, count

```python
df.groupby("continent").aggregate({"beer_servings":"std",
                                   "spirit_servings":"median",
                                   "wine_servings":["mean", "min", "max"],
                                   "total_litres_of_pure_alcohol":["mean", "count"]})
```

9- Her alkol cinsinin en çok tüketildiği ülkeyi bulun. 

```python
print(df["country"][df["beer_servings"].idxmax()] +   " is the country that drinks the most spirit.")
print(df["country"][df["spirit_servings"].idxmax()] + " is the country that drinks the most beer.")
print(df["country"][df["wine_servings"].idxmax()] +   " is the country that drinks the most vine.")
print(df["country"][df["total_litres_of_pure_alcohol"].idxmax()] +   " is the country that drinks the most pure alcohol.")
```

10- Yeni değişkenler üreterek her alkol cinsinin ilgili ülkedeki toplam alkol tüketimi içerisindeki oranlarını yazdırınız.  

```python
countries = df["country"].values
beer_rate = []
spirit_rate = []
wine_rate = []

for i in range(len(countries)):
    for x in drink_types:
        total = df[x].sum()
        if x == "beer_servings": 
            beer_rate.append(100 * (df[x][i]) / (total))
        elif x == "spirit_servings": 
            spirit_rate.append(100 * (df[x][i]) / (total))
        elif x == "wine_servings": 
            wine_rate.append(100 * (df[x][i]) / (total))
            
columns = ["country", "beer usage rate (%)", "spirit usage rate (%)", "wine usage rate (%)"]
df_rates = pd.DataFrame(data = [countries, beer_rate, spirit_rate, wine_rate]).T
df_rates.columns = columns
df_rates
```
